```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/data.csv', header=None, names=['x1', 'x2', 'y'])

X = df[['x1', 'x2']].values.astype(float)
y = df['y'].values.astype(int)

print(df.head(10))
print(df.shape)
print(df.columns.tolist())
```

```python
class0 = df[df['y'] == 0]
class1 = df[df['y'] == 1]

plt.figure(figsize=(7, 7))
plt.scatter(class0['x1'], class0['x2'], label='Class 0')
plt.scatter(class1['x1'], class1['x2'], label='Class 1')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def perceptron_part1(X, y, learning_rate=0.1, max_epochs=65, seed=42):
    rng = np.random.default_rng(seed)

    # keep random weights, but avoid tiny w2 that makes crazy lines
    while True:
        w = rng.uniform(-1, 1, size=X.shape[1])
        if abs(w[1]) > 0.2:
            break

    b = rng.uniform(-1, 1)

    lines = [(w.copy(), b)]   # initial line

    for epoch in range(max_epochs):
        errors = 0

        for xi, yi in zip(X, y):
            y_hat = 1 if (np.dot(w, xi) + b) >= 0 else 0
            error = yi - y_hat

            b = b + learning_rate * error
            w = w + learning_rate * error * xi

            if error != 0:
                errors += 1

        lines.append((w.copy(), b))   # save only once per epoch

        if errors == 0:
            break

    preds = (X @ w + b >= 0).astype(int)
    accuracy = np.mean(preds == y)

    return w, b, lines, epoch + 1, accuracy
```

```python
def plot_boundary_lines(X, y, lines, title, show_every=5):
    plt.figure(figsize=(7, 7))

    class0 = y == 0
    class1 = y == 1

    plt.scatter(X[class0, 0], X[class0, 1], label='Class 0')
    plt.scatter(X[class1, 0], X[class1, 1], label='Class 1')

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + 0.05
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    x_vals = np.array([x_min, x_max])

    used_green_label = False

    for i, (w, b) in enumerate(lines):
        if abs(w[1]) < 1e-8:
            continue

        y_vals = -(w[0] * x_vals + b) / w[1]

        if i == 0:
            plt.plot(x_vals, y_vals, 'r-', linewidth=2, label='Initial line')
        elif i == len(lines) - 1:
            plt.plot(x_vals, y_vals, 'k-', linewidth=2, label='Final line')
        elif i % show_every == 0:
            if not used_green_label:
                plt.plot(x_vals, y_vals, 'g--', alpha=0.6, label='Middle lines')
                used_green_label = True
            else:
                plt.plot(x_vals, y_vals, 'g--', alpha=0.6)

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for lr in [0.01, 0.1, 1.0]:
    w, b, lines, epochs_used, accuracy = perceptron_part1(
        X, y, learning_rate=lr, max_epochs=65, seed=42
    )

    print(f'Part 1 -> learning_rate={lr}, epochs_used={epochs_used}, accuracy={accuracy:.4f}')
    plot_boundary_lines(X, y, lines, f'Part 1 Heuristic Perceptron (lr={lr})', show_every=5)
```

```python
def sigmoid(z):
    z = np.clip(z, -500, 500)
    return 1 / (1 + np.exp(-z))

def log_loss(y_true, y_prob):
    eps = 1e-9
    y_prob = np.clip(y_prob, eps, 1 - eps)
    return -np.mean(y_true * np.log(y_prob) + (1 - y_true) * np.log(1 - y_prob))

def plot_error_curve(epoch_points, loss_points, title):
    plt.figure(figsize=(7, 5))
    plt.plot(epoch_points, loss_points, marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('Log Loss')
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
def neuron_part2(X, y, learning_rate=0.1, max_epochs=100, seed=42):
    rng = np.random.default_rng(seed)

    while True:
        w = rng.uniform(-1, 1, size=X.shape[1])
        if abs(w[1]) > 0.2:
            break

    b = rng.uniform(-1, 1)
    lines = [(w.copy(), b)]
    epoch_points = []
    loss_points = []

    for epoch in range(1, max_epochs + 1):
        for xi, yi in zip(X, y):
            z = np.dot(w, xi) + b
            y_hat = sigmoid(z)
            error = yi - y_hat

            b = b + learning_rate * error
            w = w + learning_rate * error * xi

        lines.append((w.copy(), b))

        if epoch % 10 == 0:
            probs = sigmoid(X @ w + b)
            loss = log_loss(y, probs)
            epoch_points.append(epoch)
            loss_points.append(loss)

    probs = sigmoid(X @ w + b)
    preds = (probs >= 0.5).astype(int)
    accuracy = np.mean(preds == y)

    return w, b, lines, epoch_points, loss_points, accuracy
```

```python
settings = [
    (0.01, 100),
    (0.1, 100),
    (0.1, 200)
]

for lr, epochs in settings:
    w, b, lines, epoch_points, loss_points, accuracy = neuron_part2(
        X, y, learning_rate=lr, max_epochs=epochs, seed=42
    )

    print(f'Part 2 -> learning_rate={lr}, epochs={epochs}, accuracy={accuracy:.4f}, final_loss={loss_points[-1]:.4f}')
    plot_boundary_lines(X, y, lines, f'Part 2 Neuron (lr={lr}, epochs={epochs})', show_every=5)
    plot_error_curve(epoch_points, loss_points, f'Part 2 Error Plot (lr={lr}, epochs={epochs})')
```
